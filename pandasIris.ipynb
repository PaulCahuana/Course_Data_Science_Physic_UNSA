```
import pandas as pd #librería para poder extraer datos de excel o csv, en otras palabras en tablas
import warnings #librería  para manejar excepciones
warnings.filterwarnings("ignore")
import seaborn as sns #es para mandar nuestros datos en bloques para poder graficar
import matplotlib.pyplot as plt #librería para poder graficar
sns.set(style="white", color_codes=True) 
```

```
iris = pd.read_csv("Iris.csv") 
```

```
iris.head()
```

```
iris.shape  # mostramos nuestro tamaño
```

```
iris.dtypes
```

```
iris.plot(kind="scatter",x="SepalLengthCm", y="SepalWidthCm")
plt.show()
```

```
iris.plot(kind="scatter",x="PetalLengthCm", y="PetalWidthCm")
plt.show()
```

```
iris.plot(kind="scatter",x="PetalLengthCm", y="PetalWidthCm", color="red")
plt.show()
```

```
#SepalLengthCm", y="SepalWidthCm
sns.FacetGrid(iris, hue="Species",size=5)\
.map(plt.scatter, "SepalLengthCm","SepalWidthCm")\
.add_legend()

plt.show()
```

```
#SepalLengthCm", y="SepalWidthCm
sns.FacetGrid(iris, hue="Species",size=5)\
.map(plt.scatter, "PetalLengthCm","PetalWidthCm")\
.add_legend()

plt.show()
```

```
colores= {'color':['blue','red','green']}
sns.FacetGrid(iris,hue_kws=colores, hue="Species",size=5)\
.map(plt.scatter, "SepalLengthCm","SepalWidthCm")\
.add_legend()
plt.show()
```

```
sns.boxplot(x="Species",y="PetalLengthCm",data=iris)
```

```
sns.boxplot(x="Species",y="SepalLengthCm",data=iris)
```
